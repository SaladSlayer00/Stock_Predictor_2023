# Data Collection

```python
from yahoo_fin import stock_info as si
import datetime as dt
import pandas_datareader as pdr
import pandas as pd
```

```python
ticker = 'TSLA'
start_date = dt.datetime(2010, 1, 1)
end_date = dt.datetime(2023, 1, 1)
```

```python
stock_data = si.get_quote_table(ticker)
```

```python
historical_data = si.get_data(ticker, start_date, end_date, interval='1d')
# delete ticker column
historical_data = historical_data.drop(columns=['ticker'])
historical_data
```

```python
# List of symbols you're interested in
fred_symbols = ['UNRATE', 'GDP', 'FEDFUNDS', 'CPIAUCNS', 'M2', 'DGS10', 'PCE', 'T10Y2Y', 'USROA', 'USROE', 'WTISPLC', 'HOUST', 'INDPRO', 'PAYEMS', 'BAMLH0A0HYM2', 'GS10', 'BASE', 'RIFSPPFAAD01NB', 'EXUSEU', 'NETEXP']
```

```python
fred_df = pdr.get_data_fred(fred_symbols, start_date, end_date)
fred_df
```

```python
# Join the two dataframes using DataFrame.join()
df_join = historical_data.join(fred_df, how='left')
# Rename first column as 'Date'
df_join_final = df_join.rename(columns={'Unnamed: 0': 'Date'})
df_join_final
```

```python
# I would like to fill all the NaN in columns of the dataframe with the previous value, if there is no previous value with the following value
dataset = df_join.fillna(method='ffill').fillna(method='bfill')
dataset
```

```python
# Reset index for historical_data and fred_df before creating Spark DataFrames
historical_data_reset = historical_data.reset_index()
fred_df_reset = fred_df.reset_index()
```

```python
# Rename index column to DATE 
historical_data_reset = historical_data_reset.rename(columns={'index': 'DATE'})
```

```python
fred_df_reset
```

```python
# Remove adjclose column
prova_dataset = dataset.drop(columns=['adjclose'])
```

```python
#Remove open, high, low, adjclose and volume columns
prova2_dataset = dataset.drop(columns=['open', 'high', 'low', 'adjclose'])
prova2_dataset
```

# Data Exploration

```python
import matplotlib.pyplot as plt

# Plotting stock prices
plt.figure(figsize=(14, 7))
plt.plot(dataset.index, dataset['close'], label=f'{ticker} Stock Price', color='blue')
plt.title(f'{ticker} Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

# Plot some economic indicators
economic_indicators_to_plot = ['UNRATE', 'GDP', 'FEDFUNDS']
for indicator in economic_indicators_to_plot:
    plt.figure(figsize=(14, 7))
    plt.plot(dataset.index, dataset[indicator], label=indicator, color='orange')
    plt.title(f'{indicator} Over Time')
    plt.xlabel('Date')
    plt.ylabel(indicator)
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Model Development

```python
from sklearn.model_selection import train_test_split

X = prova2_dataset.drop(columns=['close'])
y = prova2_dataset['close']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
# Random Forest
rf_model = RandomForestRegressor(n_estimators=500, max_depth=100, random_state=42)
rf_model.fit(X_train, y_train)
```

# Model Evaluation

```python
lr_predictions = lr_model.predict(X_test)
rf_predictions = rf_model.predict(X_test)
```

```python
rmse = mean_squared_error(y_test, lr_predictions, squared=False)
r2 = lr_model.score(X_test, y_test)
print(f"RMSE: {rmse}")
print(f"R-squared: {r2}")
```

```python
rmse = mean_squared_error(y_test, rf_predictions, squared=False)
r2 = rf_model.score(X_test, y_test)
print(f"RMSE: {rmse}")
print(f"R-squared: {r2}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))
plt.plot(y_test.index, y_test, label='Actual', color='blue')
plt.plot(y_test.index, lr_predictions, label='Predicted', color='red')
plt.title('Actual vs Predicted Stock Prices')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.plot(y_test.index, y_test, label='Actual', color='blue')
plt.plot(y_test.index, rf_predictions, label='Predicted', color='red')
plt.title('Actual vs Predicted Stock Prices')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()
```
